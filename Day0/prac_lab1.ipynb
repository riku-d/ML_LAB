```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Score": [85, 55, 72, 40]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

filtered_df = df[df["Score"] >= 60]
print("Filtered DataFrame:\n", filtered_df)
```

```python
def normalize(values):
    min_val = min(values)
    max_val = max(values)
    return [(x - min_val) / (max_val - min_val) for x in values]

nums = [10, 20, 30, 40]
print(normalize(nums))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
heart = pd.read_csv("heart.csv")
```

```python
print("Shape of dataset:", heart.shape)
```

```python
print("Missing values per column:\n", heart.isnull().sum())
```

```python
print("Data types:\n", heart.dtypes)
```

```python
print("Mean age:", heart["age"].mean())
```

```python
selected_columns = ["age", "sex", "cp", "restecg", "chol"]
heart_selected = heart[selected_columns]

train_data, test_data = train_test_split(
    heart_selected,
    test_size=0.25,
    random_state=42
)

print("Training set shape:", train_data.shape)
print("Testing set shape:", test_data.shape)
```
