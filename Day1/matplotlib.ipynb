```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("tips.csv")
```

```python
print(data.head(10))
```

```python
plt.scatter(data['day'], data['tip'])
plt.title("Day vs Tip (Scatter Plot)")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.plot(data['day'], data['tip'])
plt.title("Day vs Tip (Line Chart)")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.bar(data['day'], data['tip'])
plt.title("Day vs Tip (Bar Plot)")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

```python
plt.hist(data['total_bill'])
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```
