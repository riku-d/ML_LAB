```python
import pandas as pd

df = pd.read_csv("mtcars.csv")

# Display first few rows
print(df.head())

# Add new columns
df['Power_to_Weight'] = df['hp'] / df['wt']
df['Mileage_Category'] = df['mpg'].apply(lambda x: 'High' if x > 20 else 'Low')

# Export to new CSV
df.to_csv("mtcars_modified.csv", index=False)

# Read exported file
new_df = pd.read_csv("mtcars_modified.csv")
print(new_df.tail(10))
```

```python
import matplotlib.pyplot as plt

oil_change = [3, 1, 2, 2, 3, 5]
longevity = [290000, 190000, 240000, 250000, 290000, 350000]

plt.scatter(oil_change, longevity)
plt.title("Oil Change Frequency vs Engine Longevity")
plt.xlabel("Oil Changes per Year")
plt.ylabel("Engine Longevity (km)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = pd.read_csv("tips.csv")

# Scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()

# Bar plot
sns.barplot(x='day', y='tip', data=tips)
plt.show()

# Histogram
sns.histplot(tips['total_bill'], kde=True)
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Product': ['Pen', 'Book', 'Laptop', 'Phone'],
    'Category': ['Stationery', 'Stationery', 'Electronics', 'Electronics'],
    'Price': [10, 50, 60000, 30000],
    'Stock': [100, 200, 10, 15]
})

# a) Add Discount column
df['Discount'] = np.random.randint(5, 30, size=len(df))

# b) Final Price
df['Final Price'] = df['Price'] - (df['Price'] * df['Discount'] / 100)

# c) Drop Stock column
df.drop('Stock', axis=1, inplace=True)

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Algorithm': [85, 45, 90, 70, 95],
    'Data Structure': [88, 55, 92, 60, 98],
    'English': [90, 40, 85, 75, 99]
})

# Total score
df['Total'] = df[['Algorithm', 'Data Structure', 'English']].sum(axis=1)

# a) Above 80 in all subjects
above_80 = df[(df['Algorithm'] > 80) & 
              (df['Data Structure'] > 80) & 
              (df['English'] > 80)]

# b) Below 50 in any subject
below_50 = df[(df['Algorithm'] < 50) | 
              (df['Data Structure'] < 50) | 
              (df['English'] < 50)]

# c) Top 5% total score
top_5_percent = df[df['Total'] >= df['Total'].quantile(0.95)]

print("Above 80 in all subjects:\n", above_80)
print("\nBelow 50 in any subject:\n", below_50)
print("\nTop 5% students:\n", top_5_percent)
```
