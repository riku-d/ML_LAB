```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
# 1. Dataset
x_vals = [1, 2, 3, 4, 5]
y_vals = [3, 7, 9, 13, 15]
n = len(x_vals)

# 2. Calculate Sums manually (as per formula)
sum_x = sum(x_vals)
sum_y = sum(y_vals)
sum_xx = sum(x**2 for x in x_vals)
sum_xy = sum(x*y for x, y in zip(x_vals, y_vals))

# Calculate m and b
numerator_m = (n * sum_xy) - (sum_x * sum_y)
denominator_m = (n * sum_xx) - (sum_x ** 2)

m = numerator_m / denominator_m
b = (sum_y - (m * sum_x)) / n

print(f"Calculated Model: f(x) = {m}x + {b}")

# 4. Evaluate the Model
# a. Calculate predicted values (f_new)
f_new = [(m * x) + b for x in x_vals]

print("\nEvaluation:")
print(f"{'x':<5} {'y (Actual)':<12} {'f_new (Pred)':<12}")
for i in range(n):
    print(f"{x_vals[i]:<5} {y_vals[i]:<12} {f_new[i]:<12.2f}")

# b. Calculate Sum of Squared Differences
squared_diffs = sum((y_vals[i] - f_new[i]) ** 2 for i in range(n))
print(f"\nSum of Squared Differences: {squared_diffs:.4f}")

# 6. Prediction for x = 4.5
x_unknown = 4.5
y_pred_part1 = (m * x_unknown) + b
print(f"Prediction for x={x_unknown}: {y_pred_part1}")

# PLOTTING PART 1
plt.figure(figsize=(10, 5))
plt.scatter(x_vals, y_vals, color='blue', label='Actual Data Points')
plt.plot(x_vals, f_new, color='red', label=f'Regression Line: y={m:.1f}x+{b:.1f}')
plt.scatter([x_unknown], [y_pred_part1], color='green', s=100, zorder=5, label=f'Pred (x={x_unknown})')
plt.title('Part 01: Linear Regression Model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Create CSV with 100 random datapoints (40 to 50)
# We generate X values and create Y values with some noise to simulate a trend
csv_filename = 'random_data.csv'
np.random.seed(42) # For reproducibility

# Generate 100 random x values between 40 and 50
x_rand = np.random.uniform(40, 50, 100)
# Generate y values: let's assume a hidden relation y = 2x + 5 + random_noise
y_rand = (2 * x_rand) + 5 + np.random.normal(0, 2, 100)

# Save to CSV
df = pd.DataFrame({'x': x_rand, 'y': y_rand})
df.to_csv(csv_filename, index=False)
print(f"Generated {csv_filename} with 100 datapoints.")

# Read the CSV
data = pd.read_csv(csv_filename)
x_p2 = data['x'].values
y_p2 = data['y'].values
n_p2 = len(x_p2)

# Calculate m and b for Part 2
sum_x_p2 = np.sum(x_p2)
sum_y_p2 = np.sum(y_p2)
sum_xx_p2 = np.sum(x_p2 ** 2)
sum_xy_p2 = np.sum(x_p2 * y_p2)

m_p2 = ((n_p2 * sum_xy_p2) - (sum_x_p2 * sum_y_p2)) / ((n_p2 * sum_xx_p2) - (sum_x_p2 ** 2))
b_p2 = (sum_y_p2 - (m_p2 * sum_x_p2)) / n_p2

print(f"Part 2 Model: f(x) = {m_p2:.4f}x + {b_p2:.4f}")

# Calculate predicted line for plotting
f_new_p2 = (m_p2 * x_p2) + b_p2

# 2. Prediction for x = 43.5
x_unknown_p2 = 43.5
y_pred_p2 = (m_p2 * x_unknown_p2) + b_p2
print(f"Prediction for x={x_unknown_p2}: {y_pred_p2:.4f}")

# PLOTTING PART 2
plt.figure(figsize=(10, 5))
plt.scatter(x_p2, y_p2, color='gray', alpha=0.6, label='Random Data (CSV)')
plt.plot(x_p2, f_new_p2, color='orange', label='Regression Line')
plt.scatter([x_unknown_p2], [y_pred_p2], color='purple', s=100, zorder=5, label=f'Pred (x={x_unknown_p2})')
plt.title('Part 02: Regression on Random Data (40-50)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
